# First plot

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data glance
import os
if os.getcwd().endswith("notebooks"):
    os.chdir("..")

fNameECG = "data/ECGu.txt"
ECG = np.genfromtxt(fNameECG, delimiter=' ')
print(ECG)
print(ECG.shape)
```

```python
#data conversion
ECG = ECG * 0.001024  # convert to mV

num_rows = ECG.shape[0]
t = np.linspace(0, num_rows/1000, num_rows)  # time vector in seconds
```

```python
#first plotting
plt.plot(t, ECG)
plt.xlabel('Time (s)')
plt.ylabel('ECG (mV)')
plt.title('ECG Signal')
plt.xlim(0, 6.5)
plt.show()
```

## Realistic plot

Define three new variables for each record of the ECG in order to use them in a single cell with plt.show() at the end, to simplify the merging of three graphs into a single figure.

```python
#ECG I
ECG_I = ECG[:, 0]

#ECG II
ECG_II = ECG[:, 1]

#ECG III
ECG_III = ECG[:, 2]
```

Plot each new variable in a single standardized figure. The plotting of a variable includes the standards dimensions:
* black line;
* $y∈[-1;1.5]$ mV; 
* $x∈[0;6.5]$ sec; 
* major ticks on the ordonnate axis each $0.5$ mV, minor ones each $0.1$ mV; 
* major ticks on the abscissa axis each $0.25$ sec, minor ones each $0.05$ sec.
* rectangular red grid with major ($0.8$) and minor ($0.3$) widths according to the x and y ticks.

The goal is to have a clear plotting in which intra- and inter-comparisons of amplitude and time characteristics are easily possible. The chosen structure also offers excellent performance without the need for advanced hardware, as no iterative functions are used.

```python
fig, axs = plt.subplots(3, 1, figsize=(20, 12), sharex=True) # figure with 3 subplots, defined by "axs[...]"

# ECG I
axs[0].plot(t, ECG[:, 0], color='black')
axs[0].set_ylabel('ECG I (mV)')
axs[0].set_ylim(-1, 1.5)
axs[0].set_yticks(np.arange(-1, 1.5, 0.5))
axs[0].set_yticks(np.arange(-1, 1.5, 0.1), minor=True)
axs[0].grid(which='major', color='red', linestyle='-', linewidth=0.8)
axs[0].grid(which='minor', color='red', linestyle='--', linewidth=0.3)
axs[0].set_xlim(0, 6.5)

# ECG II
axs[1].plot(t, ECG[:, 1], color='black')
axs[1].set_ylabel('ECG II (mV)')
axs[1].set_ylim(-1, 1.5)
axs[1].set_yticks(np.arange(-1, 1.5, 0.5))
axs[1].set_yticks(np.arange(-1, 1.5, 0.1), minor=True)
axs[1].grid(which='major', color='red', linestyle='-', linewidth=0.8)
axs[1].grid(which='minor', color='red', linestyle='--', linewidth=0.3)
axs[1].set_xlim(0, 6.5)

# ECG III
axs[2].plot(t, ECG[:, 2], color='black')
axs[2].set_ylabel('ECG III (mV)')
axs[2].set_ylim(-1, 1.5)
axs[2].set_yticks(np.arange(-1, 1.5, 0.5))
axs[2].set_yticks(np.arange(-1, 1.5, 0.1), minor=True)
axs[2].grid(which='major', color='red', linestyle='-', linewidth=0.8)
axs[2].grid(which='minor', color='red', linestyle='--', linewidth=0.3)
axs[2].set_xlim(0, 6.5)
axs[2].set_xlabel('Time (s)')

plt.xticks(np.arange(0, t[-1], 0.25))
plt.gca().set_xticks(np.arange(0, t[-1], 0.05), minor=True)

plt.tight_layout()
plt.show()
```
